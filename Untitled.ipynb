```python
X=df.drop('diabetes',axis=1)
y=df['diabetes']
#droping the target for the features and move it to the label
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
#split 
```

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_leaf=1,min_samples_split=2)

#Gini(D) = 1 - Σ(p(i))^2
#the maximum depth of the decision tree
#the minimum number of samples allowed in a leaf node
#This parameter defines the minimum number of samples required to split an internal node
```

```python
# Cross-validation
cv_scores = cross_val_score(model, X_train, y_train, cv=5)  # 5-fold cross-validation
print("Cross-validation scores:", cv_scores)
print("Mean CV accuracy:", np.mean(cv_scores))
```

```python
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
y_train_pred=model.predict(X_train)
```

```python
y_pred.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
test_accuracy = accuracy_score(y_test, y_pred)
train_accuracy = accuracy_score(y_train, y_train_pred)
print("test Accuracy:", test_accuracy*100)
print("train Accuracy:", train_accuracy*100)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, log_loss

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Compute precision, recall, f1-score
cr = classification_report(y_test, y_pred)
print("Classification Report:")
print(cr)
# Compute Log Loss
ll = log_loss(y_test, y_pred)
print("Log Loss: ", ll)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create a decision tree classifier
dt_classifier = DecisionTreeClassifier()

# Instantiate GridSearchCV
grid_search = GridSearchCV(estimator=dt_classifier, param_grid=param_grid, cv=5,scoring='accuracy')

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Print the best parameters found
print("Best Parameters:", grid_search.best_params_)

# Print the best estimator found
print("Best Estimator:", grid_search.best_estimator_)
```
